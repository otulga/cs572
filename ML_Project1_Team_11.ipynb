```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```
df = pd.read_csv('../content/income_evaluation.csv')
```

```
df.head()
```

```
# Fixing column name and cell values
df.columns = df.columns.str.strip()
df = df.applymap(lambda x: x.strip() if type(x)==str else x)

# Fixing income feature
df['50K+'] = df.income.apply(lambda x : 1 if x != '<=50K' else 0)


# Fixing sex feature
df['sex'] = df.sex.apply(lambda x : 1 if x == 'Male' else 0)


# Fixing workclass feature /9 categories to 6/
df['workclass'].replace(['Private','Self-emp-not-inc','Local-gov','?','State-gov','Self-emp-inc','Federal-gov','Without-pay','Never-worked'],
                    ['private','self-emp-not-inc','local-gov','other','local-gov','self-emp-inc','federal-gov','other','other'],inplace=True)

# Fixing marital-status feature /7 categories to 4/
df['marital-status'].replace(['Divorced','Married-AF-spouse','Married-civ-spouse','Married-spouse-absent','Never-married','Separated','Widowed'],
                    ['divorced','married','married','seperated','never-married','seperated','seperated'],inplace=True)

# Fixing relationship feature /7 categories to 3/
df['relationship'].replace(['Husband', 'Not-in-family', 'Other-relative', 'Own-child', 'Unmarried', 'Wife'],
                    ['family', 'other', 'other', 'other', 'unmarried', 'family'],inplace=True)

# Fixing race feature /5 categories to 3/
df['race'].replace(['White', 'Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other'],
                    ['white', 'black', 'other', 'other', 'other'],inplace=True)

# Creating new feature from native-country
df['native'] = df['native-country'].apply(lambda x : 1 if x == 'United-States' else 0)


df.drop([
    'income',
    'native-country',
    'fnlwgt', 
    'education', 
    'education-num', 
    'capital-loss',
    'capital-gain'
],axis=1,inplace=True)


df.head()
```

```
dummy_columns = [
    'workclass', 
    'relationship', 
    'race',
    'marital-status']
df_train = pd.get_dummies(df, columns=dummy_columns, drop_first=True)

df_train.head()
```

```
from sklearn.model_selection import train_test_split

X = df_train.drop('50K+', axis=1)
y = df_train['50K+']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import tree

from sklearn.metrics import accuracy_score
```

```
from sklearn.model_selection import cross_val_score
```

```
svc = SVC(gamma = 0.01, C = 100)
scores_svc = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
print(scores_svc)
print(scores_svc.mean())
```
